```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from skimage import morphology
from queue import PriorityQueue
import cv2

def bresenhams_line_algorithm(start, end):
    """Generate points on a line from start to end using Bresenham's line algorithm."""
    points = []
    x0, y0 = start
    x1, y1 = end
    dx = abs(x1 - x0)
    dy = -abs(y1 - y0)
    sx = 1 if x0 < x1 else -1
    sy = 1 if y0 < y1 else -1
    err = dx + dy

    while True:
        points.append((x0, y0))
        if x0 == x1 and y0 == y1:
            break
        e2 = 2 * err
        if e2 >= dy:
            err += dy
            x0 += sx
        if e2 <= dx:
            err += dx
            y0 += sy
    return points

def can_skip(binary_image, start, end):
    """Check if a straight path exists between start and end points in the binary image."""
    for point in bresenhams_line_algorithm(start, end):
        if not binary_image[point[0], point[1]]:  # False indicates an obstacle in binary image
            return False
    return True

def optimize_path_for_straight_lines_and_90_degree_turns(binary_image, path):
    """Optimize the path to consist only of straight lines and 90-degree turns."""
    optimized_path = [path[0]]

    for i in range(1, len(path)):
        if i + 1 < len(path) and can_skip(binary_image, optimized_path[-1], path[i + 1]):
            continue  # Skip intermediate point if there's a clear line to the next point
        optimized_path.append(path[i])

    return optimized_path
def load_image_and_find_points_with_dilation(path):
    """Load the image, identify start and end points, apply dilation, and convert to a binary array."""
    image = Image.open(path)
    rgb_image = np.array(image)

    # Define the RGB values for blue (start) and yellow (end)
    blue = np.array([0, 0, 255])
    yellow = np.array([255, 255, 0])

    # Create masks for the start and end points
    blue_mask = np.all(rgb_image == blue, axis=-1)
    yellow_mask = np.all(rgb_image == yellow, axis=-1)

    # Find the start and end points
    start_point = np.argwhere(blue_mask)
    end_point = np.argwhere(yellow_mask)
    if start_point.size > 0 and end_point.size > 0:
        start = tuple(start_point[0])
        end = tuple(end_point[0])
    else:
        raise ValueError("Start or end point not found in the image.")

    # Adjust the threshold for what is considered white (traversable)
    white_threshold = 200
    binary_image = np.all(rgb_image >= white_threshold, axis=-1)

    # Include blue and yellow areas as traversable
    binary_image = binary_image | blue_mask | yellow_mask

    # Apply dilation to the binary image to smooth out thin black lines
    binary_image = morphology.binary_dilation(binary_image, selem=morphology.disk(1))

    return binary_image, start, end, rgb_image

def heuristic(a, b):
    """Calculate the Manhattan distance between two points."""
    return abs(a[0] - b[0]) + abs(a[1] - b[1])


def a_star_search(start, goal, graph):
    """Perform A* search."""
    from queue import PriorityQueue

    frontier = PriorityQueue()
    frontier.put((0, start))
    came_from = {}
    cost_so_far = {}
    came_from[start] = None
    cost_so_far[start] = 0

    while not frontier.empty():
        current = frontier.get()[1]

        if current == goal:
            break

        for next in neighbors(graph, current):
            new_cost = cost_so_far[current] + 1
            if next not in cost_so_far or new_cost < cost_so_far[next]:
                cost_so_far[next] = new_cost
                priority = new_cost + heuristic(goal, next)
                frontier.put((priority, next))
                came_from[next] = current

    return came_from, cost_so_far

def neighbors(graph, node):
    """Return the neighbors of a given node."""
    directions = [(0, 1), (0, -1), (1, 0), (-1, 0)]
    result = []
    for direction in directions:
        neighbor = (node[0] + direction[0], node[1] + direction[1])
        if 0 <= neighbor[0] < graph.shape[0] and 0 <= neighbor[1] < graph.shape[1]:
            if graph[neighbor]:
                result.append(neighbor)
    return result

def reconstruct_path(came_from, start, goal):
    """Reconstruct the path from start to goal."""
    current = goal
    path = []
    while current != start:
        path.append(current)
        current = came_from[current]
    path.append(start)
    path.reverse()
    return path


# def visualize_path_with_lines(rgb_image, path):
#     """Visualize the path on the RGB image by drawing lines between consecutive points."""
#     import cv2  # OpenCV for drawing lines

#     # Convert the image to BGR format for OpenCV
#     rgb_image_cv = cv2.cvtColor(rgb_image, cv2.COLOR_RGB2BGR)

#     # Set the color and thickness of the path
#     line_color = (0, 0, 255)  # Blue in BGR
#     line_thickness = 4

#     # Draw lines between each consecutive point
#     for i in range(len(path) - 1):
#         start_point = (path[i][1], path[i][0])  # Swap coordinates for OpenCV
#         end_point = (path[i+1][1], path[i+1][0])  # Swap coordinates for OpenCV
#         cv2.line(rgb_image_cv, start_point, end_point, line_color, line_thickness)

#     # Convert the image back to RGB format to display with matplotlib
#     rgb_image_plt = cv2.cvtColor(rgb_image_cv, cv2.COLOR_BGR2RGB)

#     plt.imshow(rgb_image_plt)
#     plt.axis('off')  # Hide axes for a cleaner image
#     plt.show()
# def visualize_centered_path_with_straight_lines(rgb_image, path, line_thickness=10, line_color=(255, 0, 0)):
#     """Visualize the path on the RGB image by drawing straight, centered lines between path points."""
#     for i in range(len(path) - 1):
#         start_point = (path[i][1], path[i][0])  # Swap coordinates for OpenCV (x, y)
#         end_point = (path[i+1][1], path[i+1][0])  # Swap coordinates for OpenCV (x, y)

#         # Draw a line on the image using the specified thickness and color
#         rgb_image = cv2.line(rgb_image, start_point, end_point, line_color, line_thickness)

#     plt.imshow(rgb_image)
#     plt.axis('off')  # Hide axes for a cleaner image
#     plt.show()

def visualize_path_with_turns(rgb_image, path, line_thickness=2, line_color=(255, 0, 0), turn_marker_color=(0, 255, 0)):
    """Visualize the path on the RGB image by drawing straight lines and marking turns."""
    if len(path) < 2:
        return  # Not enough points to visualize a path

    for i in range(1, len(path) - 1):
        prev_point = path[i - 1]
        curr_point = path[i]
        next_point = path[i + 1]

        # Draw the line segments
        cv2.line(rgb_image, (curr_point[1], curr_point[0]), (prev_point[1], prev_point[0]), line_color, line_thickness)
        cv2.line(rgb_image, (curr_point[1], curr_point[0]), (next_point[1], next_point[0]), line_color, line_thickness)

        # Check if there is a turn at the current point
        if (prev_point[0] != next_point[0]) and (prev_point[1] != next_point[1]):
            # Mark the turn with a circle
            cv2.circle(rgb_image, (curr_point[1], curr_point[0]), radius=10, color=turn_marker_color, thickness=-3)

    # Draw the first and last line segments
    cv2.line(rgb_image, (path[0][1], path[0][0]), (path[1][1], path[1][0]), line_color, line_thickness)
    cv2.line(rgb_image, (path[-2][1], path[-2][0]), (path[-1][1], path[-1][0]), line_color, line_thickness)

    # Highlight the start and end points
    cv2.circle(rgb_image, (path[0][1], path[0][0]), radius=5, color=(0, 0, 255), thickness=-1)  # Start
    cv2.circle(rgb_image, (path[-1][1], path[-1][0]), radius=5, color=(255, 255, 0), thickness=-1)  # End

    plt.imshow(rgb_image)
    plt.axis('off')
    plt.show()


# def refine_optimized_path(binary_image, path):
#     """Refine the optimized path to remove unnecessary and overlapping turns."""
#     if len(path) < 3:
#         return path

#     refined_path = [path[0]]

#     i = 0
#     while i < len(path) - 1:
#         # Look ahead for the farthest point that can be reached in a straight line
#         for j in range(len(path) - 1, i, -1):
#             if can_skip(binary_image, refined_path[-1], path[j]):
#                 # If we can directly reach a farther point, skip to it
#                 i = j
#                 break
#         refined_path.append(path[i])

#     return refined_path

# def refine_and_simplify_path(binary_image, original_path):
#     """Refine the path to remove unnecessary, overlapping, and close-by turns."""
#     if len(original_path) < 3:
#         return original_path

#     refined_path = [original_path[0]]

#     i = 0
#     while i < len(original_path) - 1:
#         max_reachable = i + 1  # Assume we can at least reach the next point
#         # Attempt to reach further points in a straight line
#         for j in range(i + 2, len(original_path)):
#             if can_skip(binary_image, refined_path[-1], original_path[j]):
#                 max_reachable = j
#             else:
#                 break  # Stop at the first unreachable point

#         # Before appending, check if we can bypass this point entirely for further optimization
#         next_point = max_reachable + 1 if max_reachable + 1 < len(original_path) else max_reachable
#         if next_point < len(original_path) and can_skip(binary_image, refined_path[-1], original_path[next_point]):
#             max_reachable = next_point

#         refined_path.append(original_path[max_reachable])
#         i = max_reachable  # Continue from the farthest reachable point

#     return refined_path
# def refine_path_with_turn_constraints(binary_image, path):
#     """Refine the path to enforce left and right turns while minimizing distance."""
#     if len(path) < 3:
#         return path

#     refined_path = [path[0]]

#     def is_straight_line(p1, p2, p3):
#         """Check if three points form a straight line."""
#         return (p1[0] - p2[0]) * (p2[1] - p3[1]) == (p2[0] - p3[0]) * (p1[1] - p2[1])

#     i = 1
#     while i < len(path) - 1:
#         current = refined_path[-1]
#         next_point = path[i]
#         lookahead = path[i + 1]

#         # Check if the next segment (next_point to lookahead) continues straight or requires a turn
#         if can_skip(binary_image, current, lookahead) and is_straight_line(current, next_point, lookahead):
#             # If we can skip to lookahead and it forms a straight line, do so
#             i += 1  # Increment to process the next segment
#         else:
#             # If a turn is required, add the next_point to the path
#             refined_path.append(next_point)

#         i += 1

#     # Ensure the last point is included
#     if refined_path[-1] != path[-1]:
#         refined_path.append(path[-1])

#     return refined_path
# def can_skip(binary_image, start, end):
#     """Check if a straight path exists between start and end points in the binary image."""
#     for point in bresenhams_line_algorithm(start, end):
#         if not binary_image[point[0], point[1]]:  # False indicates an obstacle in binary image
#             return False
#     return True
def refine_valid_path(binary_image, path):
    """Refine the path to ensure valid movements without going through walls."""
    if len(path) < 3:
        return path

    refined_path = [path[0]]
    i = 0

    while i < len(path) - 1:
        # Find the next point in the path that requires a change in direction
        next_change_idx = i + 1
        while next_change_idx < len(path) - 1:
            if can_skip(binary_image, path[i], path[next_change_idx + 1]):
                next_change_idx += 1
            else:
                break
        refined_path.append(path[next_change_idx])
        i = next_change_idx

    return refined_path
def correct_diagonal_movements(binary_image, path):
    """Correct diagonal movements in the path to ensure only 90-degree turns are made."""
    corrected_path = [path[0]]

    for i in range(1, len(path) - 1):
        prev_point = corrected_path[-1]
        curr_point = path[i]
        next_point = path[i + 1]

        # Check if the current segment is diagonal
        if abs(curr_point[0] - prev_point[0]) == abs(curr_point[1] - prev_point[1]):
            # Determine the direction of the turn
            if binary_image[curr_point[0], prev_point[1]]:
                # Turn is possible horizontally first
                turn_point = (curr_point[0], prev_point[1])
            else:
                # Turn is possible vertically first
                turn_point = (prev_point[0], curr_point[1])

            corrected_path.append(turn_point)

        corrected_path.append(curr_point)

    corrected_path.append(path[-1])  # Ensure the last point is included
    return corrected_path

# def refine_path_for_90_degree_turns(binary_image, path):
#     """Refine the path to ensure it consists only of 90-degree turns."""
#     if len(path) < 3:
#         return path

#     refined_path = [path[0]]

#     for i in range(1, len(path) - 1):
#         current = refined_path[-1]
#         next_point = path[i]
#         lookahead = path[i + 1]

#         # Determine if a straight path to the lookahead point is possible
#         if can_skip(binary_image, current, lookahead):
#             # Check if current to next_point to lookahead makes a 90-degree turn
#             if (current[0] == next_point[0] and next_point[0] != lookahead[0] and next_point[1] == lookahead[1]) or \
#                (current[1] == next_point[1] and next_point[1] != lookahead[1] and next_point[0] == lookahead[0]):
#                 # Skip the next_point if it's not contributing to a 90-degree turn
#                 continue
#         refined_path.append(next_point)

#     refined_path.append(path[-1])  # Add the goal point

#     return refined_path
# def extend_path_to_avoid_diagonals(binary_image, path):
#     """Extend the path to use only 90-degree left or right turns."""
#     if len(path) < 3:
#         return path  # No need to process if path is too short

#     extended_path = [path[0]]  # Start with the initial position

#     for i in range(1, len(path)):
#         current = extended_path[-1]
#         next_point = path[i]

#         # Check if the next movement is diagonal
#         if current[0] != next_point[0] and current[1] != next_point[1]:
#             # Choose to either go horizontally then vertically, or the opposite.
#             # Make the choice based on which direction is unobstructed.
#             horizontal_first = (current[0], next_point[1])
#             vertical_first = (next_point[0], current[1])

#             # Check which direction to take first - prioritize unobstructed paths
#             if binary_image[horizontal_first[0], horizontal_first[1]]:
#                 # If the horizontal move is possible, add the intermediate point
#                 extended_path.append(horizontal_first)
#             elif binary_image[vertical_first[0], vertical_first[1]]:
#                 # If the vertical move is possible, add the intermediate point
#                 extended_path.append(vertical_first)
#             else:
#                 # If neither are directly possible, choose the one that gets closer to the target
#                 if abs(horizontal_first[1] - next_point[1]) < abs(vertical_first[0] - next_point[0]):
#                     extended_path.append(horizontal_first)
#                 else:
#                     extended_path.append(vertical_first)

#         # Add the next point of the original path
#         extended_path.append(next_point)

#     return extended_path

# def extend_straight_paths(binary_image, path):
#     """Extend straight paths to minimize turns."""
#     if len(path) < 3:
#         return path  # No turns to optimize

#     # Initialize the refined path with the starting point
#     refined_path = [path[0]]

#     i = 0
#     while i < len(path) - 1:
#         # Look ahead for the next point where a turn is required
#         turn_point = path[i]
#         for j in range(i + 1, len(path)):
#             # Check if the path can continue straight without turning
#             if not can_skip(binary_image, turn_point, path[j]):
#                 # We've encountered an obstacle or the end of a straight path
#                 # Record the last traversable point before the turn
#                 turn_point = path[j - 1]
#                 break
#         else:
#             # If we reached the end without needing to turn, record the last point
#             turn_point = path[-1]

#         # Add the turn point to the path
#         refined_path.append(turn_point)
#         # Skip ahead in the path to the point after the turn
#         i = path.index(turn_point)

#     return refined_path



# Load the image, process it, and find start and end points with dilation applied
binary_image, start, end, rgb_image_for_visualization = load_image_and_find_points_with_dilation('/content/drive/MyDrive/SampleImages/Screenshot_20240212_182715.png')

# Perform A* search to find the shortest path
came_from, cost_so_far = a_star_search(start, end, binary_image.astype(np.bool_))

# Reconstruct the path
path = reconstruct_path(came_from, start, end)

# Optimize the path for straight lines and 90-degree turns
optimized_path = optimize_path_for_straight_lines_and_90_degree_turns(binary_image.astype(np.bool_), path)
# Use the refined path to optimize the path further
#refined_optimized_path = refine_optimized_path(binary_image.astype(np.bool_), optimized_path)
# Apply the refined and simplified path logic
# refined_and_simplified_path = refine_and_simplify_path(binary_image.astype(np.bool_), optimized_path)
# Apply the refined path logic with turn constraints
#refined_path_with_turns = refine_path_with_turn_constraints(binary_image.astype(np.bool_), optimized_path)
# Assuming binary_image is the binary representation of the maze where True indicates free space
refined_valid_path = refine_valid_path(binary_image.astype(np.bool_), optimized_path)
# Apply the diagonal movement correction to the refined path
corrected_path = correct_diagonal_movements(binary_image.astype(np.bool_), refined_valid_path)
# Apply the path refinement to extend straight paths and minimize turns
#refined_path_with_min_turns = extend_straight_paths(binary_image.astype(np.bool_), optimized_path)

# Visualize the path with minimal turns
#visualize_path_with_turns(rgb_image_for_visualization, refined_path_with_min_turns)
# Visualize the corrected path
visualize_path_with_turns(rgb_image_for_visualization, corrected_path)
# Visualize the refined path that does not go through walls
#visualize_path_with_turns(rgb_image_for_visualization, refined_valid_path)
# Use this function on the initial path from A* search before any optimization
initial_path = reconstruct_path(came_from, start, end)
#extended_path = extend_path_to_avoid_diagonals(binary_image.astype(np.bool_), initial_path)

# Now visualize the extended path with only 90-degree turns
#visualize_path_with_turns(rgb_image_for_visualization, extended_path)
# Visualize the refined path that adheres to turn constraints
#visualize_path_with_turns(rgb_image_for_visualization, refined_path_with_turns)
# Visualize the refined and simplified path on the original image
# visualize_path_with_turns(rgb_image_for_visualization, refined_and_simplified_path)
# Visualize the optimized path on the original image
# visualize_centered_path_with_straight_lines(rgb_image_for_visualization, optimized_path, line_thickness=10, line_color=(255, 0, 0))
# Visualize the path with marked turns on the original image
# visualize_path_with_turns(rgb_image_for_visualization, optimized_path)
# Visualize the refined optimized path on the original image
#visualize_path_with_turns(rgb_image_for_visualization, refined_optimized_path)
```

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from skimage import morphology
from queue import PriorityQueue
import cv2

def bresenhams_line_algorithm(start, end):
    """Generate points on a line from start to end using Bresenham's line algorithm."""
    points = []
    x0, y0 = start
    x1, y1 = end
    dx = abs(x1 - x0)
    dy = -abs(y1 - y0)
    sx = 1 if x0 < x1 else -1
    sy = 1 if y0 < y1 else -1
    err = dx + dy

    while True:
        points.append((x0, y0))
        if x0 == x1 and y0 == y1:
            break
        e2 = 2 * err
        if e2 >= dy:
            err += dy
            x0 += sx
        if e2 <= dx:
            err += dx
            y0 += sy
    return points

def can_skip(binary_image, start, end):
    """Check if a straight path exists between start and end points in the binary image."""
    for point in bresenhams_line_algorithm(start, end):
        if not binary_image[point[0], point[1]]:  # False indicates an obstacle in binary image
            return False
    return True

def optimize_path_for_straight_lines_and_90_degree_turns(binary_image, path):
    """Optimize the path to consist only of straight lines and 90-degree turns."""
    optimized_path = [path[0]]

    for i in range(1, len(path)):
        if i + 1 < len(path) and can_skip(binary_image, optimized_path[-1], path[i + 1]):
            continue  # Skip intermediate point if there's a clear line to the next point
        optimized_path.append(path[i])

    return optimized_path
def load_image_and_find_points_with_dilation(path):
    """Load the image, identify start and end points directly connected to colored areas, apply dilation."""
    image = Image.open(path)
    rgb_image = np.array(image)

    # Define RGB values for blue (start) and yellow (end) markers
    blue = np.array([0, 0, 255])
    yellow = np.array([255, 255, 0])

    # Create masks for identifying the start (blue) and end (yellow) areas
    blue_mask = np.all(rgb_image == blue, axis=-1)
    yellow_mask = np.all(rgb_image == yellow, axis=-1)

    # Adjust the threshold for what is considered white (traversable)
    white_threshold = 200
    binary_image = np.all(rgb_image >= white_threshold, axis=-1)

    # Include blue and yellow areas as traversable
    binary_image = binary_image | blue_mask | yellow_mask

    # Apply dilation to the binary image to smooth out thin black lines
    binary_image = morphology.binary_dilation(binary_image, selem=morphology.disk(1))

    # Initialize start and end points
    start = None
    end = None

    # Finding the rightmost white pixel directly above the yellow area
    if yellow_mask.any():
        yellow_top = np.min(np.where(yellow_mask)[0])
        for y in range(yellow_top - 1, -1, -1):  # Move upwards from the topmost yellow pixel
            if binary_image[y, :].any():
                row = binary_image[y, :]
                cols_with_white = np.where(row)[0]
                yellow_cols = np.where(yellow_mask[yellow_top])[0]
                intersect_cols = np.intersect1d(cols_with_white, yellow_cols)
                if intersect_cols.size > 0:
                    start = (y, intersect_cols[-1])  # Rightmost intersecting white pixel
                    break

    # Finding the rightmost white pixel directly below the blue area
    if blue_mask.any():
        blue_bottom = np.max(np.where(blue_mask)[0])
        for y in range(blue_bottom + 1, binary_image.shape[0]):  # Move downwards from the bottommost blue pixel
            if binary_image[y, :].any():
                row = binary_image[y, :]
                cols_with_white = np.where(row)[0]
                blue_cols = np.where(blue_mask[blue_bottom])[0]
                intersect_cols = np.intersect1d(cols_with_white, blue_cols)
                if intersect_cols.size > 0:
                    end = (y, intersect_cols[-1])  # Rightmost intersecting white pixel
                    break

    if start is None or end is None:
        raise ValueError("Start or end point not found in the image.")

    return binary_image.astype(bool), start, end, rgb_image

def heuristic(a, b):
    """Calculate the Manhattan distance between two points."""
    return abs(a[0] - b[0]) + abs(a[1] - b[1])


def a_star_search(start, goal, graph):
    """Perform A* search."""
    from queue import PriorityQueue

    frontier = PriorityQueue()
    frontier.put((0, start))
    came_from = {}
    cost_so_far = {}
    came_from[start] = None
    cost_so_far[start] = 0

    while not frontier.empty():
        current = frontier.get()[1]

        if current == goal:
            break

        for next in neighbors(graph, current):
            new_cost = cost_so_far[current] + 1
            if next not in cost_so_far or new_cost < cost_so_far[next]:
                cost_so_far[next] = new_cost
                priority = new_cost + heuristic(goal, next)
                frontier.put((priority, next))
                came_from[next] = current

    return came_from, cost_so_far

def neighbors(graph, node):
    """Return the neighbors of a given node."""
    directions = [(0, 1), (0, -1), (1, 0), (-1, 0)]
    result = []
    for direction in directions:
        neighbor = (node[0] + direction[0], node[1] + direction[1])
        if 0 <= neighbor[0] < graph.shape[0] and 0 <= neighbor[1] < graph.shape[1]:
            if graph[neighbor]:
                result.append(neighbor)
    return result

def reconstruct_path(came_from, start, goal):
    """Reconstruct the path from start to goal."""
    current = goal
    path = []
    while current != start:
        path.append(current)
        current = came_from[current]
    path.append(start)
    path.reverse()
    return path



def visualize_path_with_turns(rgb_image, path, line_thickness=2, line_color=(255, 0, 0), turn_marker_color=(0, 255, 0)):
    """Visualize the path on the RGB image by drawing straight lines and marking turns."""
    if len(path) < 2:
        return  # Not enough points to visualize a path

    for i in range(1, len(path) - 1):
        prev_point = path[i - 1]
        curr_point = path[i]
        next_point = path[i + 1]

        # Draw the line segments
        cv2.line(rgb_image, (curr_point[1], curr_point[0]), (prev_point[1], prev_point[0]), line_color, line_thickness)
        cv2.line(rgb_image, (curr_point[1], curr_point[0]), (next_point[1], next_point[0]), line_color, line_thickness)

        # Check if there is a turn at the current point
        if (prev_point[0] != next_point[0]) and (prev_point[1] != next_point[1]):
            # Mark the turn with a circle
            cv2.circle(rgb_image, (curr_point[1], curr_point[0]), radius=10, color=turn_marker_color, thickness=-3)

    # Draw the first and last line segments
    cv2.line(rgb_image, (path[0][1], path[0][0]), (path[1][1], path[1][0]), line_color, line_thickness)
    cv2.line(rgb_image, (path[-2][1], path[-2][0]), (path[-1][1], path[-1][0]), line_color, line_thickness)

    # Highlight the start and end points
    cv2.circle(rgb_image, (path[0][1], path[0][0]), radius=5, color=(0, 0, 255), thickness=-1)  # Start
    cv2.circle(rgb_image, (path[-1][1], path[-1][0]), radius=5, color=(255, 255, 0), thickness=-1)  # End

    plt.imshow(rgb_image)
    plt.axis('off')
    plt.show()

def refine_valid_path(binary_image, path):
    """Refine the path to ensure valid movements without going through walls."""
    if len(path) < 3:
        return path

    refined_path = [path[0]]
    i = 0

    while i < len(path) - 1:
        # Find the next point in the path that requires a change in direction
        next_change_idx = i + 1
        while next_change_idx < len(path) - 1:
            if can_skip(binary_image, path[i], path[next_change_idx + 1]):
                next_change_idx += 1
            else:
                break
        refined_path.append(path[next_change_idx])
        i = next_change_idx

    return refined_path
def correct_diagonal_movements(binary_image, path):
    """Correct diagonal movements in the path to ensure only 90-degree turns are made."""
    corrected_path = [path[0]]

    for i in range(1, len(path) - 1):
        prev_point = corrected_path[-1]
        curr_point = path[i]
        next_point = path[i + 1]

        # Check if the current segment is diagonal
        if abs(curr_point[0] - prev_point[0]) == abs(curr_point[1] - prev_point[1]):
            # Determine the direction of the turn
            if binary_image[curr_point[0], prev_point[1]]:
                # Turn is possible horizontally first
                turn_point = (curr_point[0], prev_point[1])
            else:
                # Turn is possible vertically first
                turn_point = (prev_point[0], curr_point[1])

            corrected_path.append(turn_point)

        corrected_path.append(curr_point)

    corrected_path.append(path[-1])  # Ensure the last point is included
    return corrected_path


# Load the image, process it, and find start and end points with dilation applied
binary_image, start, end, rgb_image_for_visualization = load_image_and_find_points_with_dilation('/content/drive/MyDrive/SampleImages/Screenshot_20240212_182933.png')

# Perform A* search to find the shortest path
came_from, cost_so_far = a_star_search(start, end, binary_image.astype(np.bool_))

# Reconstruct the path
path = reconstruct_path(came_from, start, end)

# Optimize the path for straight lines and 90-degree turns
optimized_path = optimize_path_for_straight_lines_and_90_degree_turns(binary_image.astype(np.bool_), path)

refined_valid_path = refine_valid_path(binary_image.astype(np.bool_), optimized_path)
# Apply the diagonal movement correction to the refined path
corrected_path = correct_diagonal_movements(binary_image.astype(np.bool_), refined_valid_path)

visualize_path_with_turns(rgb_image_for_visualization, corrected_path)
initial_path = reconstruct_path(came_from, start, end)
```

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from skimage import morphology
from queue import PriorityQueue
import cv2

def bresenhams_line_algorithm(start, end):
    """Generate points on a line from start to end using Bresenham's line algorithm."""
    points = []
    x0, y0 = start
    x1, y1 = end
    dx = abs(x1 - x0)
    dy = -abs(y1 - y0)
    sx = 1 if x0 < x1 else -1
    sy = 1 if y0 < y1 else -1
    err = dx + dy

    while True:
        points.append((x0, y0))
        if x0 == x1 and y0 == y1:
            break
        e2 = 2 * err
        if e2 >= dy:
            err += dy
            x0 += sx
        if e2 <= dx:
            err += dx
            y0 += sy
    return points

def can_skip(binary_image, start, end):
    """Check if a straight path exists between start and end points in the binary image."""
    for point in bresenhams_line_algorithm(start, end):
        if not binary_image[point[0], point[1]]:  # False indicates an obstacle in binary image
            return False
    return True

def optimize_path_for_straight_lines_and_90_degree_turns(binary_image, path):
    """Optimize the path to consist only of straight lines and 90-degree turns."""
    optimized_path = [path[0]]

    for i in range(1, len(path)):
        if i + 1 < len(path) and can_skip(binary_image, optimized_path[-1], path[i + 1]):
            continue  # Skip intermediate point if there's a clear line to the next point
        optimized_path.append(path[i])

    # Refine the optimized path to directly connect to the next turn
    refined_path = [optimized_path[0]]
    for i in range(1, len(optimized_path)):
        prev_point = refined_path[-1]
        curr_point = optimized_path[i]
        if i + 1 < len(optimized_path):
            next_point = optimized_path[i + 1]
            # Check if the next point forms a straight line with the current and previous points
            if (next_point[0] == curr_point[0] == prev_point[0]) or (next_point[1] == curr_point[1] == prev_point[1]):
                continue  # Skip adding the current point if it forms a straight line
        # Add the current point to the refined path
        refined_path.append(curr_point)

    return refined_path

def load_image_and_find_points_with_dilation(path):
    """Load the image, identify start and end points directly connected to colored areas, apply dilation."""
    image = Image.open(path)
    rgb_image = np.array(image)

    # Define RGB values for blue (start) and yellow (end) markers
    blue = np.array([0, 0, 255])
    yellow = np.array([255, 255, 0])

    # Create masks for identifying the start (blue) and end (yellow) areas
    blue_mask = np.all(rgb_image == blue, axis=-1)
    yellow_mask = np.all(rgb_image == yellow, axis=-1)

    # Adjust the threshold for what is considered white (traversable)
    white_threshold = 200
    binary_image = np.all(rgb_image >= white_threshold, axis=-1)

    # Include blue and yellow areas as traversable
    binary_image = binary_image | blue_mask | yellow_mask

    # Apply dilation to the binary image to smooth out thin black lines
    binary_image = morphology.binary_dilation(binary_image, selem=morphology.disk(1))

    # Initialize start and end points
    start = None
    end = None

    # Finding the rightmost white pixel directly above the yellow area
    if yellow_mask.any():
        yellow_top = np.min(np.where(yellow_mask)[0])
        for y in range(yellow_top - 1, -1, -1):  # Move upwards from the topmost yellow pixel
            if binary_image[y, :].any():
                row = binary_image[y, :]
                cols_with_white = np.where(row)[0]
                yellow_cols = np.where(yellow_mask[yellow_top])[0]
                intersect_cols = np.intersect1d(cols_with_white, yellow_cols)
                if intersect_cols.size > 0:
                    start = (y, intersect_cols[-1])  # Rightmost intersecting white pixel
                    break

    # Finding the rightmost white pixel directly below the blue area
    if blue_mask.any():
        blue_bottom = np.max(np.where(blue_mask)[0])
        for y in range(blue_bottom + 1, binary_image.shape[0]):  # Move downwards from the bottommost blue pixel
            if binary_image[y, :].any():
                row = binary_image[y, :]
                cols_with_white = np.where(row)[0]
                blue_cols = np.where(blue_mask[blue_bottom])[0]
                intersect_cols = np.intersect1d(cols_with_white, blue_cols)
                if intersect_cols.size > 0:
                    end = (y, intersect_cols[-1])  # Rightmost intersecting white pixel
                    break

    if start is None or end is None:
        raise ValueError("Start or end point not found in the image.")

    return binary_image.astype(bool), start, end, rgb_image

def heuristic(a, b):
    """Calculate the Manhattan distance between two points."""
    return abs(a[0] - b[0]) + abs(a[1] - b[1])


def a_star_search(start, goal, graph):
    """Perform A* search."""
    from queue import PriorityQueue

    frontier = PriorityQueue()
    frontier.put((0, start))
    came_from = {}
    cost_so_far = {}
    came_from[start] = None
    cost_so_far[start] = 0

    while not frontier.empty():
        current = frontier.get()[1]

        if current == goal:
            break

        for next in neighbors(graph, current):
            new_cost = cost_so_far[current] + 1
            if next not in cost_so_far or new_cost < cost_so_far[next]:
                cost_so_far[next] = new_cost
                priority = new_cost + heuristic(goal, next)
                frontier.put((priority, next))
                came_from[next] = current

    return came_from, cost_so_far

def neighbors(graph, node):
    """Return the neighbors of a given node."""
    directions = [(0, 1), (0, -1), (1, 0), (-1, 0)]
    result = []
    for direction in directions:
        neighbor = (node[0] + direction[0], node[1] + direction[1])
        if 0 <= neighbor[0] < graph.shape[0] and 0 <= neighbor[1] < graph.shape[1]:
            if graph[neighbor]:
                result.append(neighbor)
    return result

def reconstruct_path(came_from, start, goal):
    """Reconstruct the path from start to goal."""
    current = goal
    path = []
    while current != start:
        path.append(current)
        current = came_from[current]
    path.append(start)
    path.reverse()
    return path

def visualize_path_with_turns(rgb_image, path, line_thickness=2, line_color=(255, 0, 0), turn_marker_color=(0, 255, 0)):
    """Visualize the path on the RGB image by drawing straight lines and marking turns."""
    if len(path) < 2:
        return rgb_image  # No path to visualize, return the original image

    image_copy = rgb_image.copy()  # Make a copy of the original image

    for i in range(1, len(path) - 1):
        prev_point = path[i - 1]
        curr_point = path[i]
        next_point = path[i + 1]

        # Draw the line segments
        cv2.line(image_copy, (curr_point[1], curr_point[0]), (prev_point[1], prev_point[0]), line_color, line_thickness)
        cv2.line(image_copy, (curr_point[1], curr_point[0]), (next_point[1], next_point[0]), line_color, line_thickness)

        # Check if there is a turn at the current point
        if (prev_point[0] != next_point[0]) and (prev_point[1] != next_point[1]):
            # Mark the turn with a circle
            cv2.circle(image_copy, (curr_point[1], curr_point[0]), radius=0, color=turn_marker_color, thickness=-1)

    # Draw the first and last line segments
    cv2.line(image_copy, (path[0][1], path[0][0]), (path[1][1], path[1][0]), line_color, line_thickness)
    cv2.line(image_copy, (path[-2][1], path[-2][0]), (path[-1][1], path[-1][0]), line_color, line_thickness)

    # Highlight the start and end points
    cv2.circle(image_copy, (path[0][1], path[0][0]), radius=5, color=(0, 0, 255), thickness=-1)  # Start
    cv2.circle(image_copy, (path[-1][1], path[-1][0]), radius=5, color=(255, 255, 0), thickness=-1)  # End

    return image_copy

def refine_valid_path(binary_image, path):
    """Refine the path to ensure valid movements without going through walls."""
    if len(path) < 3:
        return path

    refined_path = [path[0]]
    i = 0

    while i < len(path) - 1:
        # Find the next point in the path that requires a change in direction
        next_change_idx = i + 1
        while next_change_idx < len(path) - 1:
            if can_skip(binary_image, path[i], path[next_change_idx + 1]):
                next_change_idx += 1
            else:
                break
        refined_path.append(path[next_change_idx])
        i = next_change_idx

    return refined_path

def correct_diagonal_movements(binary_image, path):
    """Correct diagonal movements in the path to ensure only 90-degree turns are made."""
    corrected_path = [path[0]]

    for i in range(1, len(path) - 1):
        prev_point = corrected_path[-1]
        curr_point = path[i]
        next_point = path[i + 1]

        # Check if the current segment is diagonal
        if abs(curr_point[0] - prev_point[0]) == abs(curr_point[1] - prev_point[1]):
            # Determine the direction of the turn
            if binary_image[curr_point[0], prev_point[1]]:
                # Turn is possible horizontally first
                turn_point = (curr_point[0], prev_point[1])
            else:
                # Turn is possible vertically first
                turn_point = (prev_point[0], curr_point[1])

            corrected_path.append(turn_point)

        corrected_path.append(curr_point)

    corrected_path.append(path[-1])  # Ensure the last point is included
    return corrected_path

# Load the image, process it, and find start and end points with dilation applied
binary_image, start, end, rgb_image_for_visualization = load_image_and_find_points_with_dilation('/content/drive/MyDrive/SampleImages/Screenshot_20240212_183051.png')

# Perform A* search to find the shortest path
came_from, cost_so_far = a_star_search(start, end, binary_image.astype(np.bool_))

# Reconstruct the path
path = reconstruct_path(came_from, start, end)

# Optimize the path for straight lines and 90-degree turns
optimized_path = optimize_path_for_straight_lines_and_90_degree_turns(binary_image.astype(np.bool_), path)

# Refine the path
refined_valid_path = refine_valid_path(binary_image.astype(np.bool_), optimized_path)

# Apply the diagonal movement correction to the refined path
corrected_path = correct_diagonal_movements(binary_image.astype(np.bool_), refined_valid_path)

# Visualize the path on the original RGB image
visualized_image = visualize_path_with_turns(rgb_image_for_visualization, corrected_path)

# Save the visualized image
output_image_path = '/content/drive/MyDrive/SampleImages/visualized_image.png'
plt.imshow(visualized_image)
plt.axis('off')
plt.savefig(output_image_path)

# Save the original RGB image
output_original_image_path = '/content/drive/MyDrive/SampleImages/original_image.png'
plt.imshow(rgb_image_for_visualization)
plt.axis('off')
plt.savefig(output_original_image_path)

# Output the path to the console
print(f"Visualized image saved to {output_image_path}")
print(f"Original RGB image saved to {output_original_image_path}")
from PIL import Image
import numpy as np
from skimage import morphology
import cv2

# Paste your existing code here

def get_turn_directions_and_distances(path):
    """Get turn directions and distances."""
    directions = []
    distances = []
    last_turn_index = 0

    for i in range(1, len(path) - 1):
        prev = path[i - 1]
        curr = path[i]
        next = path[i + 1]

        dx1, dy1 = curr[0] - prev[0], curr[1] - prev[1]
        dx2, dy2 = next[0] - curr[0], next[1] - curr[1]

        cross_product = dx1 * dy2 - dy1 * dx2

        if cross_product != 0:
            direction = 'left' if cross_product > 0 else 'right'
            directions.append(direction)

            distance = sum(abs(path[last_turn_index][j] - path[i][j]) for j in range(2))
            distances.append(distance)

            last_turn_index = i

    distances.append(sum(abs(path[last_turn_index][j] - path[-1][j]) for j in range(2)))
    return directions, distances

def count_highlighted_turns_in_path(path):
    """Count the number of highlighted turns in the path."""
    if len(path) < 3:
        return 0

    highlighted_turns = 0

    for i in range(1, len(path) - 1):
        prev_point = path[i - 1]
        curr_point = path[i]
        next_point = path[i + 1]

        if (prev_point[0] != next_point[0]) and (prev_point[1] != next_point[1]):
            highlighted_turns += 1

    return highlighted_turns

def save_path_to_txt_file(path, file_name):
    """Save the path to a .txt file."""
    directions, distances = get_turn_directions_and_distances(path)
    num_turns = count_highlighted_turns_in_path(path)

    with open(file_name, 'w') as f:
        f.write(f"{num_turns}\n")
        f.write(f"{sum(distances)}\n")
        for direction in directions:
            f.write(f"{direction}\n")

# Paste the remaining part of your code here
# Save the path to a text file
output_file_name = '/content/drive/MyDrive/SampleImages/path.txt'
save_path_to_txt_file(corrected_path, output_file_name)

print(f"Path saved to {output_file_name}")
```

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from skimage import morphology
from queue import PriorityQueue
import cv2

def bresenhams_line_algorithm(start, end):
    """Generate points on a line from start to end using Bresenham's line algorithm."""
    points = []
    x0, y0 = start
    x1, y1 = end
    dx = abs(x1 - x0)
    dy = -abs(y1 - y0)
    sx = 1 if x0 < x1 else -1
    sy = 1 if y0 < y1 else -1
    err = dx + dy

    while True:
        points.append((x0, y0))
        if x0 == x1 and y0 == y1:
            break
        e2 = 2 * err
        if e2 >= dy:
            err += dy
            x0 += sx
        if e2 <= dx:
            err += dx
            y0 += sy
    return points

def can_skip(binary_image, start, end):
    """Check if a straight path exists between start and end points in the binary image."""
    for point in bresenhams_line_algorithm(start, end):
        if not binary_image[point[0], point[1]]:  # False indicates an obstacle in binary image
            return False
    return True

def optimize_path_for_straight_lines_and_90_degree_turns(binary_image, path):
    """Optimize the path to consist only of straight lines and 90-degree turns."""
    optimized_path = [path[0]]

    for i in range(1, len(path)):
        if i + 1 < len(path) and can_skip(binary_image, optimized_path[-1], path[i + 1]):
            continue  # Skip intermediate point if there's a clear line to the next point
        optimized_path.append(path[i])

    return optimized_path


from skimage.morphology import binary_opening

def apply_global_smoothing(binary_image, opening_radius=1):
    """Apply morphological opening to the entire binary image for global smoothing."""
    selem = morphology.disk(opening_radius)
    smoothed_image = binary_opening(binary_image, selem)
    return smoothed_image


def load_image_and_find_points_with_dilation(path):
    """Load the image, identify start and end points directly connected to colored areas, apply dilation."""
    image = Image.open(path)
    rgb_image = np.array(image)

    # Define RGB values for blue (start) and yellow (end) markers
    blue = np.array([0, 0, 255])
    yellow = np.array([255, 255, 0])

    # Create masks for identifying the start (blue) and end (yellow) areas
    blue_mask = np.all(rgb_image == blue, axis=-1)
    yellow_mask = np.all(rgb_image == yellow, axis=-1)

    # Adjust the threshold for what is considered white (traversable)
    white_threshold = 200
    binary_image = np.all(rgb_image >= white_threshold, axis=-1)

    # Include blue and yellow areas as traversable
    binary_image = binary_image | blue_mask | yellow_mask

    # Apply dilation to the binary image to smooth out thin black lines
    binary_image = morphology.binary_dilation(binary_image, selem=morphology.disk(1))
    # Apply global smoothing
    binary_image = apply_global_smoothing(binary_image, opening_radius=2)
    # Enhanced smoothing near markers
    binary_image = dilate_marker_areas(binary_image, blue_mask, yellow_mask, dilation_radius=3)

    # Initialize start and end points
    start = None
    end = None

    # Finding the rightmost white pixel directly above the yellow area
    if yellow_mask.any():
        yellow_top = np.min(np.where(yellow_mask)[0])
        for y in range(yellow_top - 1, -1, -1):  # Move upwards from the topmost yellow pixel
            if binary_image[y, :].any():
                row = binary_image[y, :]
                cols_with_white = np.where(row)[0]
                yellow_cols = np.where(yellow_mask[yellow_top])[0]
                intersect_cols = np.intersect1d(cols_with_white, yellow_cols)
                if intersect_cols.size > 0:
                    start = (y, intersect_cols[-1])  # Rightmost intersecting white pixel
                    break

    # Finding the rightmost white pixel directly below the blue area
    if blue_mask.any():
        blue_bottom = np.max(np.where(blue_mask)[0])
        for y in range(blue_bottom + 1, binary_image.shape[0]):  # Move downwards from the bottommost blue pixel
            if binary_image[y, :].any():
                row = binary_image[y, :]
                cols_with_white = np.where(row)[0]
                blue_cols = np.where(blue_mask[blue_bottom])[0]
                intersect_cols = np.intersect1d(cols_with_white, blue_cols)
                if intersect_cols.size > 0:
                    end = (y, intersect_cols[-1])  # Rightmost intersecting white pixel
                    break

    if start is None or end is None:
        raise ValueError("Start or end point not found in the image.")

    return binary_image.astype(bool), start, end, rgb_image

from skimage import morphology

def dilate_marker_areas(binary_image, blue_mask, yellow_mask, dilation_radius=3):
    """Apply additional dilation to the areas around the blue and yellow markers."""
    # Create a structuring element for dilation
    selem = morphology.disk(dilation_radius)

    # Dilate only the regions of the blue and yellow masks
    blue_dilated = morphology.binary_dilation(blue_mask, selem)
    yellow_dilated = morphology.binary_dilation(yellow_mask, selem)

    # Combine the dilated areas with the original binary image
    binary_image |= blue_dilated | yellow_dilated
    return binary_image


def heuristic(a, b):
    """Calculate the Manhattan distance between two points."""
    return abs(a[0] - b[0]) + abs(a[1] - b[1])


def a_star_search(start, goal, graph):
    """Perform A* search."""
    from queue import PriorityQueue

    frontier = PriorityQueue()
    frontier.put((0, start))
    came_from = {}
    cost_so_far = {}
    came_from[start] = None
    cost_so_far[start] = 0

    while not frontier.empty():
        current = frontier.get()[1]

        if current == goal:
            break

        for next in neighbors(graph, current):
            new_cost = cost_so_far[current] + 1
            if next not in cost_so_far or new_cost < cost_so_far[next]:
                cost_so_far[next] = new_cost
                priority = new_cost + heuristic(goal, next)
                frontier.put((priority, next))
                came_from[next] = current

    return came_from, cost_so_far

def neighbors(graph, node):
    """Return the neighbors of a given node."""
    directions = [(0, 1), (0, -1), (1, 0), (-1, 0)]
    result = []
    for direction in directions:
        neighbor = (node[0] + direction[0], node[1] + direction[1])
        if 0 <= neighbor[0] < graph.shape[0] and 0 <= neighbor[1] < graph.shape[1]:
            if graph[neighbor]:
                result.append(neighbor)
    return result

def reconstruct_path(came_from, start, goal):
    """Reconstruct the path from start to goal."""
    current = goal
    path = []
    while current != start:
        path.append(current)
        current = came_from[current]
    path.append(start)
    path.reverse()
    return path

def visualize_path_with_turns(rgb_image, path, line_thickness=2, line_color=(255, 0, 0), turn_marker_color=(0, 255, 0)):
    """Visualize the path on the RGB image by drawing straight lines and accurately marking turns, excluding the start point."""
    if len(path) < 3:
        return  # Not enough points to visualize a path accurately

    # Draw the first line segment without marking it as a turn
    cv2.line(rgb_image, (path[0][1], path[0][0]), (path[1][1], path[1][0]), line_color, line_thickness)

    # Iterate over the path points to draw lines and determine turns, starting from the second point
    for i in range(1, len(path) - 1):
        # Previous, current, and next points in the path
        prev_point = path[i - 1]
        curr_point = path[i]
        next_point = path[i + 1]

        # Draw line segment from current to next point
        cv2.line(rgb_image, (curr_point[1], curr_point[0]), (next_point[1], next_point[0]), line_color, line_thickness)

        # Determine if there's an actual turn at the current point, excluding the start point
        is_actual_turn = ((prev_point[0] == curr_point[0] and curr_point[1] == next_point[1]) or
                          (prev_point[1] == curr_point[1] and curr_point[0] == next_point[0])) and \
                          (prev_point[0] != next_point[0] and prev_point[1] != next_point[1])

        if is_actual_turn:
            # Mark the turn with a circle
            cv2.circle(rgb_image, (curr_point[1], curr_point[0]), radius=5, color=turn_marker_color, thickness=-1)

    plt.imshow(rgb_image)
    plt.axis('off')
    plt.show()

def visualize_path_with_turns_and_no_endpoints(rgb_image, path, line_thickness=2, line_color=(255, 0, 0), turn_marker_color=(0, 255, 0)):
    """Visualize the path on the RGB image by drawing straight lines and marking turns, excluding start and end points."""
    if len(path) < 2:
        return  # Not enough points to visualize a path

    # Draw the line segments between path points
    for i in range(1, len(path)):
        prev_point = path[i - 1]
        curr_point = path[i]
        cv2.line(rgb_image, (prev_point[1], prev_point[0]), (curr_point[1], curr_point[0]), line_color, line_thickness)

    # Mark the turns with a specific color
    for i in range(1, len(path) - 1):
        prev_point = path[i - 1]
        curr_point = path[i]
        next_point = path[i + 1]

        # A turn is identified by a change in direction
        if (prev_point[0] != next_point[0]) and (prev_point[1] != next_point[1]):
            cv2.circle(rgb_image, (curr_point[1], curr_point[0]), radius=5, color=turn_marker_color, thickness=-1)

    plt.imshow(rgb_image)
    plt.axis('off')
    plt.show()
def visualize_path_with_turns_corrected(rgb_image, path, line_thickness=2, line_color=(255, 0, 0), turn_marker_color=(0, 255, 0), skip_turns_near_start=10):
    """Visualize the path on the RGB image by drawing straight lines and accurately marking turns, with an option to skip turns near the start."""
    if len(path) < 3:
        return  # Not enough points to visualize a path accurately

    # Draw all line segments
    for i in range(1, len(path)):
        cv2.line(rgb_image, (path[i-1][1], path[i-1][0]), (path[i][1], path[i][0]), line_color, line_thickness)

    # Mark turns, skipping the specified number of initial segments
    for i in range(1, len(path) - 1):
        # Skip marking turns for the initial segments as specified
        if i <= skip_turns_near_start:
            continue

        prev_point = path[i - 1]
        curr_point = path[i]
        next_point = path[i + 1]

        # Determine if there's an actual turn at the current point
        if (prev_point[0] != next_point[0]) and (prev_point[1] != next_point[1]):
            # Both segments form a 90-degree angle
            if ((curr_point[0] - prev_point[0]) != 0 and (curr_point[1] - next_point[1]) != 0) or \
               ((curr_point[1] - prev_point[1]) != 0 and (curr_point[0] - next_point[0]) != 0):
                cv2.circle(rgb_image, (curr_point[1], curr_point[0]), radius=5, color=turn_marker_color, thickness=-1)

    plt.imshow(rgb_image)
    plt.axis('off')
    plt.show()



def enforce_straight_lines_and_90_degree_turns(path):
    if len(path) < 3:
        return path  # No simplification needed for paths with less than 3 points

    simplified_path = [path[0]]  # Always start with the first point

    for i in range(1, len(path) - 1):
        prev_point = simplified_path[-1]
        curr_point = path[i]
        next_point = path[i + 1]

        # Check if current direction is horizontal or vertical
        horizontal = curr_point[1] != prev_point[1]
        vertical = curr_point[0] != prev_point[0]

        # Predict next direction
        next_horizontal = next_point[1] != curr_point[1]
        next_vertical = next_point[0] != curr_point[0]

        # If direction changes, check if it's a necessary 90-degree turn
        if (horizontal and next_vertical) or (vertical and next_horizontal):
            # If the turn is towards the next point, keep it
            simplified_path.append(curr_point)
        elif (horizontal == next_horizontal) and (vertical == next_vertical):
            # If continuing in the same direction, skip this point to straighten the path
            continue  # Skip adding curr_point to simplified_path
        else:
            # For any other scenario, add the current point (safeguard)
            simplified_path.append(curr_point)

    simplified_path.append(path[-1])  # Ensure the last point is always included

    return simplified_path


def extend_path_to_avoid_diagonals(binary_image, path):
    """Extend the path to use only 90-degree left or right turns."""
    if len(path) < 3:
        return path  # No need to process if path is too short

    extended_path = [path[0]]  # Start with the initial position

    for i in range(1, len(path)):
        current = extended_path[-1]
        next_point = path[i]

        # Check if the next movement is diagonal
        if current[0] != next_point[0] and current[1] != next_point[1]:
            # Choose to either go horizontally then vertically, or the opposite.
            # Make the choice based on which direction is unobstructed.
            horizontal_first = (current[0], next_point[1])
            vertical_first = (next_point[0], current[1])

            # Check which direction to take first - prioritize unobstructed paths
            if binary_image[horizontal_first[0], horizontal_first[1]]:
                # If the horizontal move is possible, add the intermediate point
                extended_path.append(horizontal_first)
            elif binary_image[vertical_first[0], vertical_first[1]]:
                # If the vertical move is possible, add the intermediate point
                extended_path.append(vertical_first)
            else:
                # If neither are directly possible, choose the one that gets closer to the target
                if abs(horizontal_first[1] - next_point[1]) < abs(vertical_first[0] - next_point[0]):
                    extended_path.append(horizontal_first)
                else:
                    extended_path.append(vertical_first)

        # Add the next point of the original path
        extended_path.append(next_point)

    return extended_path

def correct_diagonal_movements_strict(binary_image, path):
    """
    Strictly correct diagonal movements in the path to ensure only 90-degree turns are made,
    by introducing intermediate points if necessary.
    """
    corrected_path = [path[0]]

    for i in range(1, len(path)):
        prev_point = corrected_path[-1]
        curr_point = path[i]

        # Direct horizontal or vertical movement is allowed
        if prev_point[0] == curr_point[0] or prev_point[1] == curr_point[1]:
            corrected_path.append(curr_point)
        else:
            # Diagonal movement detected, introduce an intermediate point
            # Choose the intermediate point based on binary_image traversability
            horizontal_first = (prev_point[0], curr_point[1])
            vertical_first = (curr_point[0], prev_point[1])

            # Prefer horizontal movement first if traversable
            if binary_image[horizontal_first[0], horizontal_first[1]]:
                corrected_path.append(horizontal_first)
            elif binary_image[vertical_first[0], vertical_first[1]]:
                corrected_path.append(vertical_first)
            else:
                # If neither direct path is traversable, keep the original point
                # This case may need special handling depending on your scenario
                print("Warning: Adjusted path may not be fully traversable.")

            corrected_path.append(curr_point)

    return corrected_path

# Load the image, process it, and find start and end points with dilation applied
binary_image, start, end, rgb_image_for_visualization = load_image_and_find_points_with_dilation('/content/drive/MyDrive/SampleImages/testimage.png')

# Perform A* search to find the shortest path
came_from, cost_so_far = a_star_search(start, end, binary_image.astype(np.bool_))

# Reconstruct the path
path = reconstruct_path(came_from, start, end)

# Optimize the path for straight lines and 90-degree turns
optimized_path = optimize_path_for_straight_lines_and_90_degree_turns(binary_image.astype(np.bool_), path)
initial_path = reconstruct_path(came_from, start, end)
extended_path = extend_path_to_avoid_diagonals(binary_image.astype(np.bool_), initial_path)

# Now visualize the extended path with only 90-degree turns
visualize_path_with_turns_corrected(rgb_image_for_visualization, extended_path)
#visualize_path_with_turns_and_no_endpoints(rgb_image_for_visualization, extended_path)
def show_binary_image(binary_image):
    """Show the binary image."""
    plt.figure(figsize=(6, 6))
    plt.title("Binary Image")
    plt.imshow(binary_image, cmap='gray')
    plt.axis('off')
    plt.show()
show_binary_image(binary_image)
```
